### Question 3: 

What are the changes of Ethereum before and after the merge? Please identify at least three different changes to elaborate on. 

Please provide references that could back-up your answer. 

Please insert Figures/Tables to better represent your idea. 





The Merge of Ethereum leads to changes in consensus mechanism, ETH issuance, and energy expenditure.

*   **Consensus Mechanism:** *Figure 2* shows that the consensus mechanism of Ethereum changed from proof-of-work (PoW) to proof-of-stake (PoS) ([“The Merge” 2022](https://ethereum.org/en/upgrades/merge/#main-content)).

  ![Illustration of The Merge](https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/Yutong-Quan/main/image/Illustration%20of%20The%20Merge.png)
  *Figure 2. Illustration of The Merge ([“The Merge” 2022](https://ethereum.org/en/upgrades/merge/#main-content))*

*   **ETH Issuance:** The issuance of ETH is the process of creating ETH that did not previously exist. Before the merge, miners were issued approximately 13,000 ETH/day, and stakers are issued approximately 1,700 ETH/day, for a total of 14,700 ETH/day. Since the merge, only the ~1,700 ETH/day remains, dropping total new ETH issuance by ~88% ([“How the Merge Impacted ETH Supply” 2022](https://ethereum.org/en/upgrades/merge/issuance/)). *Figure 3* provides a visual illustration of the change in daily block rewards before and after the merge.

  ![Ethereum Daily Block Rewards Chart](https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/Yutong-Quan/main/image/ethereum-daily-block-rew.jpeg)
  *Figure 3. Ethereum Daily Block Rewards Chart ([Etherscan 2022](https://etherscan.io/chart/blockreward))*

*   **Energy Expenditure:** When it was using PoW consensus mechanism, the Ethereum network yielded an estimate of 78 TWh for the network’s annual electricity consumption. After the merge, the energy consumption of the Ethereum network using PoS model drops to 0.0026 TWh/yr ([“Ethereum Energy Consumption” 2022](https://ethereum.org/en/energy-consumption/)). The Merge reduced Ethereum's energy consumption by ~99.95% ([“The Merge” 2022](https://ethereum.org/en/upgrades/merge/#main-content)). *Figure 4* shows an interesting comparison of Ethereum's energy expenditure before and after the merge.

  ![Ethereum Energy Consumption Pre and Post The Merge](https://raw.githubusercontent.com/Rising-Stars-by-Sunshine/Yutong-Quan/main/image/energy_consumption_pre_post_merge.png)
  *Figure 4. Comparison of Energy Consumption of Ethereum "pre-Merge" and "post-Merge" ([“Ethereum Energy Consumption” 2022](https://ethereum.org/en/energy-consumption/))*
